# Calculate Cottrell equation

```python
from softpotato import *
import softpotato as sp
```

### Define parameters:

```python
n = 1 # number of electrons
A = 1 # cm2, electrode area
CO = 1e-6 # mol/cm3, bulk concentration of O
DO = 1e-5 # cm2/s, diffusion coefficient of O

Es = -0.5 # V, step potential
ttot = 1 # s, total time
dt = 0.01 # s, time increment
```

### Generate potential waveform

Although we don't need an array of E to calculate the current according to the Cottrell equation, we call the Step method here to generate the time array that is needed.

```python
wf = sp.tech.Step(Es=Es, ttot=ttot, dt=dt)
t = wf.t
```

### Calculate Cottrell equation:

```python
macro = sp.calc.Macro(n=n, A=A, DO=DO, CO=CO)
i = macro.Cottrell(t)

import numpy as np
tsqrt = 1/np.sqrt(t)

sp.plot.plot(t, i*1e3, xlab='$t$ / s', ylab='$i$ / mA', fig=1)
sp.plot.plot(tsqrt, i*1e3, xlab='$t^{-1/2}$ / s$^{-1/2}$', ylab='$i$ / mA', mark='-o', fig=2)
```

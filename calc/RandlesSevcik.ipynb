# Calculate Randles-Sevcik equation

```python
from softpotato import *
import softpotato as sp
```

### Define parameters:

```python
n = 1 # number of electrons
A = 1 # cm2, electrode area
CO = 1e-6 # mol/cm3, bulk concentration of O
DO = 1e-5 # cm2/s, diffusion coefficient of O
T = 298 # K, temperature

import numpy as np
sr = np.linspace(0.01, 0.1, 10) # V/s, scan rate
print(sr)
```

### Calculate Randles-Sevcik

```python
macro = sp.calc.Macro(n=n, A=A, DO=DO, CO=CO, T=T)
i = macro.RandlesSevcik(sr)

sp.plot.plot(np.sqrt(sr), i*1e3, xlab=r'$\nu^{1/2}$ / V$^{1/2}$ s$^{-1/2}$', ylab='$i$ / mA', mark='o-', fig=1)
```
